```python
import pandas as pd
import numpy  as np
import ast
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

## feature selection

```python
movies = movies[['movie_id', 'genres', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.head(1)
```

## data cleaning

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'][0]
```

```python
def extract(obj):
    ls = []
    for i in ast.literal_eval(obj):
        ls.append(i['name'])
    return ls
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies['keywords'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
movies['cast'][0]
```

```python
def extract_member(obj):
    ls = []
    count = 0
    for x in ast.literal_eval(obj):
        if count != 3:
            ls.append(x['name'])
            count += 1
        else:
            break
    return ls
```

```python
movies['cast'] = movies['cast'].apply(extract_member)
```

```python
movies['crew'][0]
```

```python
def extract_dir(obj):
    ls = []
    for x in ast.literal_eval(obj):
        if x['job'] == 'Director':
            ls.append(x['name'])
            break
    return ls
```

```python
movies['crew'] = movies['crew'].apply(extract_dir)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(5)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(5)
```

```python
movies['tags'] = movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
```

```python
movies.head(1)
```

```python
movies_data = movies[['movie_id','title','tags']]
```

```python
movies_data.head(5)
```

```python
movies_data['tags'] = movies_data['tags'].apply(lambda x: " ".join(x))
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(text):
    ls = []
    for i in text.split():
        ls.append(ps.stem(i))
    return " ".join(ls)
```

```python
movies_data['tags'].apply(stemming)
```

```python
cv = CountVectorizer()
vector = cv.fit_transform(movies_data['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
feature_names = cv.get_feature_names_out()
```

```python
feature_names = list(cv.vocabulary_.keys())
```

```python
feature_names
```

## finding similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[0]
```

```python
movies_data[movies_data['title'] == 'Batman Begins'].index[0]
```

```python
def recommend(movie):
    index = movies_data[movies_data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_data.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
